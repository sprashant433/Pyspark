```python
def executeSQL(
            dbConnectType='',
            spark = None,
            DDL='No',
            sqlString=''
            
    ):
        """
        Function which accepts arguments as listed below to ...
        and returns as specified below
 
        Args:
                dbConnectType : dbConnectType as argument takes string value it is the type of connection
                sqlString='' : sqlString as argument takes string args defined by default empty string it takes database table name
 
        Returns:
                returns the df where the properties sql jdbc driver connections
        """
        #self.logMessage(f'''executeSQL:\n{sqlString} ''')
        try:
            if dbConnectType.lower() == "pyspark" or dbConnectType.lower() == "hadoop":
                df = spark.sql(sqlString)
            elif dbConnectType.lower() == "snowflake":
                sfURL = self.config.get(f'{dbConnectType}.sfURL')
                sfUser = self.config.get(f'{dbConnectType}.sfUser')
                sfPassword = self.config.get(f'{dbConnectType}.sfPassword')
                sfDatabase = self.config.get(f'{dbConnectType}.sfDatabase')
                sfSchema = self.config.get(f'{dbConnectType}.sfSchema')
                sfWarehouse = self.config.get(f'{dbConnectType}.sfWarehouse')
                               
                self.sfOptions  = {
                "sfUrl" : sfURL,
                "sfUser": sfUser,
                "sfPassword": sfPassword,
                "sfWarehouse" : sfWarehouse,
                "sfDatabase" : sfDatabase,
                "sfSchema" : sfSchema,
                "insecuremode" : "true"
                }
                df = spark.read \
                    .format("net.snowflake.spark.snowflake") \
                    .options(**self.sfOptions) \
                    .option("query", sqlString) \
                    .load()
                # spark._jvm.net.snowflake.spark.snowflake.Utils.runQuery(
                #         self.snowflake_options, query)
            elif dbConnectType.lower() =='bigquery': #any other jdbc connection

                if DDL.lower()=='yes':
                    bqclient = AuthBq()
                    try:
                        print("Connection established properly!!")
                        query_job = bqclient.query(sqlString)
                        print("Query Execution done!!")
                    except:
                        print("Connection is not established properly!!")
                else:
                    df = spark.read \
                          .format("bigquery") \
                          .option("table", table) \
                          .load()
                    return df
               
            else: #any other jdbc connection
                DB_DRIVER = self.config.get(f'{dbConnectType}.driver')
                JDBC_URL = self.config.get(f'{dbConnectType}.url')
                DB_USER = self.config.get(f'{dbConnectType}.user')
                DB_PASS = self.config.get(f'{dbConnectType}.password')
 
                df = spark.read.format("jdbc") \
                    .option("driver", DB_DRIVER) \
                    .option("url", JDBC_URL) \
                    .option("user", DB_USER) \
                    .option("password", DB_PASS) \
                    .option("query", sqlString) \
                    .load()
            # df = df.cache()
            # df.show(5)
            # df = query_job.to_dataframe()
            
            # return df
        except Exception as ex:
            self.logError(f'Error - SQL EXCEPTION: {str(ex)}', None)
            return None
```

```python
import pandas as pd

def parseInputColumn2(dsd, dlm, clen, remaining, ctype, doubleDelimiter=False):
    """
    Function which accepts arguments as listed below to ...
    and returns as specified below

    Args:
        dsd:

        dlm:

        clen:

        remaining:

        ctype:

    Returns:
        returns
    """
    try:

        result = None
        if dsd and remaining.startswith('"'):
            j = remaining[1:].index('"')
            result = remaining[1:j - 1].rstrip().replace(dlm, ' ')
            remaining = remaining[j:]
        else:
            i = remaining.find(dlm)
            if i == -1:
                i = len(remaining)
            i = min(i, clen)
            result = remaining[0:i].rstrip()
            remaining = remaining[i + 1:]
        remaining = remaining.lstrip()
        if ctype == 'str':
            result = f''' '{result}' '''
        return result, remaining

    except Exception as ex:

        print(f"Exception: {ex}")

        return

```
