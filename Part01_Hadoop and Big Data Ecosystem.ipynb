# HADOOP - Complete Revision Notes for Interview

---

## 1. HADOOP HISTORY & EVOLUTION ⏳

**Timeline:**
- **2002**: GFS (Google File System) - Distributed file system by Google
- **2004**: GMR (Google MapReduce) - Computation framework for GFS
- **2005-2006**: Hadoop created by Doug Cutting & Mike Cafarella
  - HDFS (Hadoop Distributed File System)
  - MapReduce (MR) for distributed computing
  - Primary Language: JAVA

**Hadoop Versions:**
- **0.x & 1.x**: Same architecture (NameNode, DataNode, JobTracker, TaskTracker)
- **2.x & 3.x**: Introduced YARN (Yet Another Resource Negotiator)
  - Better resource management
  - Overcomes limitations of MapReduce-only architecture

---

## 2. HADOOP ARCHITECTURE 🏗️

### Core Components:
1. **HDFS (Storage Layer)**
2. **MapReduce (Processing Layer)**
3. **YARN (Resource Management Layer)** - from Hadoop 2.x

### 5 Daemon Processes in Hadoop:
**HDFS Daemons:**
- **JP1**: NameNode (Master)
- **JP2**: DataNode (Slave)
- **JP3**: Secondary NameNode (Checkpoint Node)

**MapReduce Daemons:**
- **JP4**: JobTracker (Master) - in Hadoop 1.x
- **JP5**: TaskTracker (Slave) - in Hadoop 1.x

**In Hadoop 2.x+:**
- JobTracker → ResourceManager + ApplicationMaster
- TaskTracker → NodeManager

---

## 3. HDFS (Hadoop Distributed File System) 📁

### What is a File System?
- Used to read/write data from/to Hard Disk
- Examples: NTFS (Windows), EXT (Linux), HDFS, S3

### Types of File Systems:
1. **Standalone**: NTFS, EXT4 (single machine)
2. **Distributed**: HDFS, S3, CFS (across multiple machines)

### Blocks in HDFS:
**What is a Block?**
- Big files divided into small chunks
- NTFS: 16KB
- EXT: 512KB
- **HDFS Block Size:**
  - Hadoop 0.x & 1.x: **64MB**
  - Hadoop 2.x & 3.x: **128MB**
  - **Configurable** based on data volume

**Why Large Block Size in HDFS?**
- Reduces metadata overhead in NameNode
- Minimizes seek time
- Optimized for large files
- Better sequential read performance

### Replication in HDFS:
- **Default Replication Factor: 3**
- Creates copies of each block for fault tolerance
- Replicas stored on different DataNodes
- **Storage Cost**: 1GB data = 3GB storage (with RF=3)
- **Configurable** (can be 0, 1, 2, 3, or more)

**Interview Q: What about Storage Cost?**
- Answer: Replication is configurable but reducing it compromises fault tolerance
- Better approach: Use affordable commodity hardware
- Trade-off between cost and reliability
- Cloud storage (S3, Azure Blob) offers built-in replication

---

## 4. DISTRIBUTED SYSTEMS ARCHITECTURE 🌐

### Types of Distributed Systems:

**1. Master-Slave Architecture:**
- Examples: Hadoop, Spark, Storm
- **Advantages:**
  - Centralized control
  - Easier management
  - Simple coordination
- **Disadvantages:**
  - **SPOF** (Single Point of Failure)
  - **SPOC** (Single Point of Communication)
  - If Master goes down, entire cluster fails

**Solution to SPOF in Hadoop 2.x+:**
- High Availability (HA) with Active/Standby NameNode
- ZooKeeper for automatic failover

**2. Peer-to-Peer Architecture:**
- Example: Cassandra (NoSQL DB)
- No single point of failure
- All nodes are equal

---

## 5. NAMENODE (Master) 🧠

**Responsibilities:**
- Stores metadata (file names, locations, permissions)
- Manages file system namespace
- Tracks which blocks are on which DataNode
- Handles client read/write requests
- Performs block replication and deletion

**Metadata Storage:**
- **FSImage**: Snapshot of file system metadata
- **EditLog**: Transaction log of changes
- Stored in memory for fast access

**Secondary NameNode (Checkpoint Node):**
- NOT a backup of NameNode
- Merges FSImage and EditLog periodically
- Reduces NameNode startup time
- Creates checkpoints

---

## 6. DATANODE (Slave) 💾

**Responsibilities:**
- Stores actual data blocks
- Serves read/write requests from clients
- Performs block creation, deletion, replication
- Sends heartbeat to NameNode (every 3 seconds by default)
- Block reports (every 6 hours by default)

**Heartbeat Mechanism:**
- If NameNode doesn't receive heartbeat for 10 minutes:
  - DataNode considered dead
  - NameNode initiates re-replication of blocks

---

## 7. RACK AWARENESS POLICY 🏢

**What is a Rack?**
- Collection of DataNodes in same physical location
- Connected to same network switch

**Default Replication Strategy (RF=3):**
- **1st Replica**: Same node where client is located (or random if client is outside cluster)
- **2nd Replica**: Different rack (for rack-level fault tolerance)
- **3rd Replica**: Same rack as 2nd replica but different node

**Why Rack Awareness?**
- Fault tolerance against rack failure
- Network bandwidth optimization
- Balances reliability and performance

---

## 8. MAPREDUCE FRAMEWORK 🗺️

**What is MapReduce?**
- Programming model for processing large datasets
- Works in two phases: Map and Reduce

**Phases:**
1. **Map Phase**: 
   - Input data → Key-Value pairs
   - Filtering, sorting, transformation
2. **Shuffle & Sort**: 
   - Groups same keys together
3. **Reduce Phase**: 
   - Aggregates values for each key
   - Final output

**Components (Hadoop 1.x):**
- **JobTracker** (Master):
  - Schedules tasks
  - Monitors task execution
  - Re-executes failed tasks
- **TaskTracker** (Slave):
  - Executes Map/Reduce tasks
  - Sends progress to JobTracker

**Limitations of MapReduce:**
- Only batch processing (not real-time)
- High latency
- Not suitable for iterative algorithms
- Tight coupling with YARN in Hadoop 1.x

---

## 9. YARN (Yet Another Resource Negotiator) 🧵

**Introduced in Hadoop 2.x**

**Components:**
1. **ResourceManager (RM)** - Master
   - Global resource scheduler
   - Allocates resources to applications
2. **NodeManager (NM)** - Slave
   - Per-node resource manager
   - Monitors containers
3. **ApplicationMaster (AM)**
   - Per-application master
   - Negotiates resources from RM
   - Works with NM to execute tasks
4. **Container**
   - Encapsulation of resources (CPU, memory, disk)
   - Execution environment for tasks

**Benefits:**
- Multi-tenancy (run multiple frameworks)
- Better resource utilization
- Scalability
- Not limited to MapReduce

---

## 10. HADOOP ECOSYSTEM 🌳

### Abstraction of MapReduce (these need MR to work):
1. **HIVE**
   - Created by Facebook
   - SQL interface for Hadoop (HiveQL)
   - Query engine, NOT a database
   - Converts SQL to MapReduce jobs
   - Use case: Data warehousing, analytics
   - Supports ACID transactions (from Hive 0.14+)

2. **PIG**
   - Created by Yahoo
   - Scripting language: PIG Latin
   - Dataflow language
   - Alternative to Java for MapReduce
   - Use case: ETL operations

3. **SQOOP**
   - SQL to Hadoop
   - Import/Export between RDBMS and Hadoop
   - Command-line tool
   - Use case: Data pipeline, ETL
   - Supports incremental imports

4. **OOZIE**
   - Created by Yahoo
   - Workflow scheduler for Hadoop
   - XML-based workflow definition
   - Supports MapReduce, Pig, Hive, Sqoop jobs
   - Dependency management

### Independent Components (don't need MR):

5. **HBase**
   - Hadoop Database (NoSQL)
   - Created by Facebook
   - Column-oriented database
   - Real-time read/write access
   - Built on top of HDFS
   - Use case: Random real-time access to Big Data

6. **Mahout**
   - Machine Learning library
   - Scalable ML algorithms
   - Classification, clustering, recommendation
   - Use case: Data science jobs

7. **Flume**
   - Data ingestion tool
   - Collects, aggregates, moves log data
   - Streaming data into HDFS
   - Message Queue (MQ)
   - **Note**: Only retrieves data (one-way)
   - Alternatives: Kafka, Scribe

8. **ZooKeeper**
   - Coordination service
   - Configuration management
   - Naming service
   - Distributed synchronization
   - Used for HA in Hadoop

9. **Kafka** (often used with Hadoop)
   - Distributed streaming platform
   - Pub-sub messaging
   - Real-time data pipelines
   - Two-way communication (unlike Flume)

## 11. LOOSELY COUPLED FRAMEWORK 🔗

**What is Loosely Coupled?**
- Components are independent
- Removing one component doesn't break others
- Modular architecture

**Hadoop is Loosely Coupled:**
- Can remove Pig, Hive, Sqoop without affecting HDFS
- Can use HDFS without MapReduce
- Can integrate third-party tools easily

**Benefits:**
- Flexibility
- Easy maintenance
- Easier upgrades
- Framework interoperability

## 13. IMPORTANT INTERVIEW CONCEPTS 💡

### Write-Once, Read-Many (WORM):
- HDFS optimized for write once, read many times
- Not suitable for random writes
- Append operation supported (from Hadoop 2.x)

### Data Locality:
- "Move computation to data, not data to computation"
- MapReduce tasks run on nodes where data resides
- Reduces network congestion
- Improves performance

### Speculative Execution:
- Runs duplicate tasks if one is slow (straggler)
- First completed task's output is used
- Others are killed
- Improves job completion time

### Combiner:
- Mini-reducer at Map side
- Reduces data transfer to Reducer
- Optimization technique
- Not always applicable

### Partitioner:
- Determines which Reducer gets which key
- Default: HashPartitioner
- Custom partitioners possible
- Controls load balancing

### InputFormat & OutputFormat:
- **InputFormat**: How to read input data
  - TextInputFormat, SequenceFileInputFormat
- **OutputFormat**: How to write output
  - TextOutputFormat, SequenceFileOutputFormat

### Small Files Problem:
- HDFS not optimized for millions of small files
- Each file creates metadata in NameNode
- Memory overhead
- **Solutions**:
  - HAR (Hadoop Archive)
  - Sequence Files
  - CombineFileInputFormat
  - Merge files before loading

### Safemode:
- NameNode startup mode
- Read-only state
- Checks block replication
- Won't allow writes until safe mode is off

## 14. FAULT TOLERANCE IN HADOOP 🛡️

### DataNode Failure:
- Heartbeat mechanism detects failure
- NameNode re-replicates under-replicated blocks
- Automatic recovery

### NameNode Failure (Hadoop 1.x):
- SPOF - entire cluster goes down
- Manual intervention needed
- Recovery from Secondary NameNode

### NameNode HA (Hadoop 2.x+):
- Active-Standby NameNode configuration
- ZooKeeper for automatic failover
- Shared edit log (QJM - Quorum Journal Manager)
- Zero downtime

### TaskTracker/NodeManager Failure:
- JobTracker/ResourceManager reschedules tasks
- Tasks re-executed on other nodes

## 15. PERFORMANCE TUNING ⚡

### MapReduce Tuning:

### Data Serialization:

- Better than CSV/JSON
- Schema evolution support

## 17. COMMON INTERVIEW QUESTIONS 🎤

**Q1: Why is Hadoop 2.x better than Hadoop 1.x?**
- YARN separates resource management from processing
- NameNode HA eliminates SPOF
- Not limited to MapReduce
- Better scalability and resource utilization

**Q2: What happens when NameNode goes down?**
- **Hadoop 1.x**: Entire cluster unavailable (SPOF)
- **Hadoop 2.x+**: Automatic failover to Standby NameNode (HA)

**Q3: How does Hadoop handle node failures?**
- Heartbeat mechanism (DataNode → NameNode every 3 sec)
- After 10 min timeout: node marked dead
- Automatic re-replication of blocks

**Q4: What is the role of Secondary NameNode?**
- Merges FSImage and EditLog (checkpointing)
- NOT a backup NameNode
- Reduces NameNode restart time

**Q5: When to use Hadoop vs Spark?**
- **Hadoop**: Batch processing, cost-effective, large-scale ETL
- **Spark**: Real-time/near real-time, iterative algorithms, in-memory processing

**Q6: How to handle small files in Hadoop?**
- Use HAR (Hadoop Archive)
- Sequence Files
- Merge files before ingestion
- Use CombineFileInputFormat

**Q7: Explain Data Locality in MapReduce**
- Tasks execute on nodes where data resides
- 3 levels: Node-local → Rack-local → Off-rack
- Minimizes network I/O

**Q8: What is Speculative Execution?**
- Launches duplicate tasks for stragglers
- First completed task's output used
- Improves overall job time

**Q9: Difference between HDFS block and OS block?**
- OS block: 4KB-512KB
- HDFS block: 128MB (default)
- HDFS blocks can span multiple OS blocks

**Q10: Can we change replication factor of existing files?**
- Yes, using: `hdfs dfs -setrep -w 2 /path/to/file`
- NameNode adjusts replication accordingly

## 18. HDFS SHELL COMMANDS 💻

```bash

hdfs dfs -ls /path

hdfs dfs -mkdir /user/data

hdfs dfs -put localfile.txt /user/data/

hdfs dfs -mv /source /destination

hdfs dfs -rm /path/file.txt

# Delete directory recursively
hdfs dfs -rm -r /path/directory

# View file content
hdfs dfs -cat /path/file.txt

# View first n lines
hdfs dfs -head /path/file.txt

# View last n lines
hdfs dfs -tail /path/file.txt

# Check file size
hdfs dfs -du -h /path

# Check disk usage
hdfs dfs -df -h /

hdfs dfs -setrep -w 2 /path/file.txt

# Check file system
```

# ETL vs ELT: A Complete Explanation

## Layman's Terms

Imagine you're moving to a new house and need to organize your belongings:

**ETL (Extract, Transform, Load)** is like:
- **Extract**: Collecting items from your old house
- **Transform**: Sorting, cleaning, and packing them in a storage unit before moving
- **Load**: Delivering the already-organized boxes to your new house

You do all the organizing work *before* things arrive at the destination.

**ELT (Extract, Load, Transform)** is like:
- **Extract**: Collecting items from your old house
- **Load**: Moving everything directly to your new house (even if messy)
- **Transform**: Sorting and organizing once everything is already there

You move first, organize later.

**Why does this matter?** 
- ETL is better when your new house (database) is small - you don't want to clutter it with unsorted stuff
- ELT is better when your new house is massive with lots of space (like a warehouse) - you can dump everything and organize it there

---

## Technical Terms

### **ETL (Extract, Transform, Load)**

**Process:**
1. **Extract**: Pull raw data from source systems (databases, APIs, files)
2. **Transform**: Process data in a staging area/ETL server (clean, validate, aggregate, join, filter)
3. **Load**: Insert transformed data into the target data warehouse

**Example:**
```
Source: E-commerce Database
↓
[ETL Tool: Talend/Informatica]
├─ Extract: Pull customer orders from MySQL
├─ Transform: 
│  ├─ Remove duplicate orders
│  ├─ Calculate total revenue per customer
│  ├─ Join with customer demographics
│  └─ Aggregate by region
↓
Load: Insert into Data Warehouse (Oracle/Teradata)
```

**Characteristics:**
- Transformation happens on a separate ETL server
- Reduced load on target warehouse
- Legacy approach, predates cloud data warehouses
- Better for structured data with known transformations

---

### **ELT (Extract, Load, Transform)**

**Process:**
1. **Extract**: Pull raw data from sources
2. **Load**: Load raw/minimally processed data directly into target
3. **Transform**: Use the data warehouse's compute power to transform data

**Example:**
```
Source: E-commerce Database
↓
[Modern Data Platform]
├─ Extract: Pull raw customer orders
├─ Load: Dump raw data into Snowflake/BigQuery
↓
Transform (using SQL in warehouse):
├─ CREATE TABLE cleaned_orders AS 
│  SELECT DISTINCT customer_id, order_date, amount
│  FROM raw_orders WHERE amount > 0
└─ CREATE TABLE customer_summary AS
   SELECT customer_id, SUM(amount) as total_revenue
   FROM cleaned_orders GROUP BY customer_id
```

**Characteristics:**
- Raw data lands in warehouse first
- Transformation uses warehouse's native compute (SQL/Spark)
- Modern approach for cloud data warehouses (Snowflake, BigQuery, Redshift)
- Flexible - can re-transform without re-extracting
- Better for semi-structured data (JSON, XML)

---

## Key Differences in Data Pipeline

| Aspect | ETL | ELT |
|--------|-----|-----|
| **When transforms happen** | Before loading (external server) | After loading (in warehouse) |
| **Best for** | Traditional on-premise warehouses | Cloud data warehouses |
| **Data types** | Structured data | Structured + semi-structured |
| **Flexibility** | Less flexible (must re-run full pipeline) | More flexible (raw data available) |
| **Compute power** | Separate ETL tool/server | Leverages warehouse's scalability |
| **Time to data** | Slower (transform first) | Faster (load first, transform later) |
| **Tools** | Informatica, Talend, SSIS | dbt, Matillion, Fivetran + SQL |

---

## Real-World Scenario

**Business need:** Analyze daily sales data from 50 retail stores

**ETL Approach:**
```
9:00 PM: Extract sales from stores → 
9:30 PM: Transform on ETL server (clean, aggregate) → 
11:00 PM: Load into warehouse → 
11:30 PM: Analysts can query
```

**ELT Approach:**
```
9:00 PM: Extract sales from stores → 
9:15 PM: Load raw data into Snowflake → 
9:20 PM: Analysts can start basic queries
9:30 PM: Transformed tables ready for advanced analytics
```

The modern trend is **ELT** because cloud warehouses like Snowflake and BigQuery are so powerful they can handle transformations faster than traditional ETL tools, and you keep the raw data for future flexibility.

# ACID Properties of Databases

## Layman's Terms

Think of ACID like the rules that keep your **bank account safe and accurate**:

**Atomicity** - "All or Nothing"
- Like transferring money: Either the full amount leaves your account AND arrives in the other account, or nothing happens at all
- You never end up in a weird state where money disappeared from your account but never reached the destination

**Consistency** - "Rules Must Always Be Followed"
- Like account balance rules: Your balance can never go below $0 (if that's the rule)
- The database won't let you break the rules, even by accident

**Isolation** - "No Interference"
- Like two people using separate ATMs at the same time: They don't mess up each other's transactions
- Even if 100 people are accessing the database simultaneously, they don't see each other's incomplete work

**Durability** - "Permanent Once Confirmed"
- Like a receipt after payment: Once you see "Transaction Complete," it's saved forever, even if the power goes out
- Your data won't disappear due to crashes

---

## Technical Terms

### **A - Atomicity**

**Definition:** A transaction is treated as a single, indivisible unit - either all operations succeed or all fail.

**Technical Implementation:**
- Uses transaction logs (Write-Ahead Logging)
- Rollback mechanism to undo partial changes
- All-or-nothing guarantee

**Example:**
```sql
BEGIN TRANSACTION;
  UPDATE accounts SET balance = balance - 100 WHERE user_id = 1;
  UPDATE accounts SET balance = balance + 100 WHERE user_id = 2;
COMMIT;
```

**Scenarios:**
- ✅ **Success**: Both updates complete → Transaction commits
- ❌ **Failure**: If second update fails (user 2 doesn't exist) → Entire transaction rolls back
- 🚫 **Impossible**: Money deducted from user 1 but not added to user 2

---

### **C - Consistency**

**Definition:** A transaction brings the database from one valid state to another, maintaining all defined rules, constraints, and triggers.

**Technical Implementation:**
- Enforces integrity constraints (PRIMARY KEY, FOREIGN KEY, UNIQUE, CHECK)
- Validates data types
- Executes triggers and cascades
- Maintains business rules

**Example:**
```sql
-- Constraint: Account balance must be >= 0
CREATE TABLE accounts (
  user_id INT PRIMARY KEY,
  balance DECIMAL(10,2) CHECK (balance >= 0)
);

-- This transaction will FAIL (violates consistency)
UPDATE accounts SET balance = balance - 500 
WHERE user_id = 1 AND balance = 100;
-- Error: CHECK constraint violated
```

**What Consistency Prevents:**
- Orphaned records (foreign key violations)
- Duplicate primary keys
- Invalid data states
- Business rule violations

---

### **I - Isolation**

**Definition:** Concurrent transactions execute independently without interfering with each other; intermediate states of one transaction are invisible to others.

**Technical Implementation:**
- Uses locking mechanisms (row-level, table-level)
- Multi-Version Concurrency Control (MVCC)
- Isolation levels (Read Uncommitted, Read Committed, Repeatable Read, Serializable)

**Example Problem Without Isolation:**
```
Time | Transaction A              | Transaction B
-----|----------------------------|---------------------------
T1   | Read balance = $100        |
T2   |                            | Read balance = $100
T3   | Deduct $50 (balance = $50) |
T4   |                            | Deduct $30 (balance = $70)
T5   | Write balance = $50        |
T6   |                            | Write balance = $70
Result: Lost Update! Should be $20, but shows $70
```

**With Isolation:**
```
Transaction B must wait until Transaction A completes
Final balance correctly shows $20
```

**Isolation Levels:**

| Level | Dirty Read | Non-Repeatable Read | Phantom Read |
|-------|------------|---------------------|--------------|
| Read Uncommitted | ❌ Yes | ❌ Yes | ❌ Yes |
| Read Committed | ✅ No | ❌ Yes | ❌ Yes |
| Repeatable Read | ✅ No | ✅ No | ❌ Yes |
| Serializable | ✅ No | ✅ No | ✅ No |

---

### **D - Durability**

**Definition:** Once a transaction is committed, it remains permanent even in the event of system failure, power loss, or crashes.

**Technical Implementation:**
- Write-Ahead Logging (WAL)
- Transaction logs written to persistent storage
- Checkpointing mechanisms
- Recovery procedures using logs
- Replication and backups

**Example:**
```sql
BEGIN TRANSACTION;
  INSERT INTO orders (order_id, customer_id, amount) 
  VALUES (12345, 567, 99.99);
COMMIT; -- ← Once this returns success, data is durable

-- Even if server crashes 1 second later,
-- order 12345 will still exist after restart
```

**How It Works:**
1. Transaction changes written to log file (on disk)
2. Log is flushed to permanent storage
3. COMMIT returns success to user
4. Later, changes are written to actual data files
5. If crash occurs, database replays the log on restart

---

## ACID in Action: Real-World Example

### **Online Shopping Checkout**

```sql
BEGIN TRANSACTION;

-- Step 1: Deduct from inventory
UPDATE inventory 
SET quantity = quantity - 1 
WHERE product_id = 101;

-- Step 2: Create order
INSERT INTO orders (customer_id, product_id, amount) 
VALUES (5001, 101, 49.99);

-- Step 3: Charge payment
INSERT INTO payments (order_id, amount, status) 
VALUES (LAST_INSERT_ID(), 49.99, 'completed');

-- Step 4: Update customer points
UPDATE customers 
SET loyalty_points = loyalty_points + 50 
WHERE customer_id = 5001;

COMMIT;
```

**How ACID Protects This:**

- **Atomicity**: If payment fails, the inventory returns to original quantity, order isn't created, and loyalty points aren't added
- **Consistency**: Can't sell negative inventory, can't create order for non-existent customer (foreign key), amount must be positive
- **Isolation**: If two customers try to buy the last item simultaneously, only one succeeds - no overselling
- **Durability**: After "Order Confirmed" message, even if server crashes, your order is safe

## ACID vs BASE (NoSQL Comparison)

| Property | ACID (SQL) | BASE (NoSQL) |
|----------|------------|--------------|
| **Guarantee** | Strong consistency | Eventual consistency |
| **Best for** | Financial systems, inventory | Social media, analytics |
| **Example DBs** | PostgreSQL, MySQL, Oracle | MongoDB, Cassandra, DynamoDB |
| **Trade-off** | Slower, less scalable | Faster, highly scalable |

## When ACID Matters Most

✅ **Critical Use Cases:**
- Banking and financial transactions
- E-commerce order processing
- Healthcare records
- Inventory management
- Booking systems (flights, hotels)

❌ **Less Critical:**
- Social media likes/views counters
- Analytics and logging
- Content recommendation systems
- Sensor data collection

**Bottom Line:** ACID properties ensure your database is reliable, accurate, and trustworthy - essential when data correctness is non-negotiable.
